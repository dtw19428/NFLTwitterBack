```
# import relevant packages
import requests
from pprint import pprint
import json
import csv
import pandas as pd
```

```
# establish api target connection details
api_key = "dcec1dab9c934b97871b493a87165e6a" #Note: Enter user-specific API key
year_id = str(2020) #Note: enter specific year of interest
api_url = "https://api.sportsdata.io/v3/nfl/scores/json/Schedules/" + year_id+ "?key=" + api_key

# ping api and convert response to json format
response = requests.get(api_url)
data     = response.json()

#check json output
pprint(data[0])
```

```
# convert json format to csv
  
# open a file for writing 
schedule = open('nfl_schedule_2020.csv', 'w') 
  
# create the csv writer object 
csv_writer = csv.writer(schedule) 
  
# Counter variable used for writing  
# headers to the CSV file 
count = 0
  
for game in data: 
    if count == 0: 
  
        # Writing headers of CSV file 
        header = game.keys() 
        csv_writer.writerow(header) 
        count += 1
  
    # Writing data of CSV file 
    csv_writer.writerow(game.values()) 
  
schedule.close() 
```
