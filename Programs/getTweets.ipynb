```python
def getTweets(playerID):
    tweets = []
    player_Data = []
    for status in tweepy.Cursor(api.user_timeline, screen_name=playerID).items():
        player_Data.append(status)
    for tweet in player_Data:
        tweetjson = json.dumps(tweet._json)
        parsedtweetjson = json.loads(tweetjson)
        tweets.append([parsedtweetjson['id'],parsedtweetjson['user']['followers_count'],parsedtweetjson['favorite_count'],parsedtweetjson['retweet_count'],parsedtweetjson['created_at']])
    return tweets
```

```python
def getPlayersTweets(playerIDList):
    for index, row in playerIDList.iterrows():
        playertweets= []
        playerhandle = row['Twitter Handle']
        try:
            playertweets = getTweets(playerhandle)
        except tweepy.TweepError:
            logging.warning("taking a nap while processing " + str(playerhandle))
            time.sleep(900)
            playertweets = getTweets(playerhandle)            
        fileName = str("tweets/"+ str(row['Sportsdata.Io ID']) + "_tweets"+".csv")
        with open(fileName, 'w', newline='', encoding='utf-8') as file:
            write = csv.writer(file)
            write.writerows(playertweets)
```

```python
from pprint import pprint
import tweepy, csv, json, time, logging
import pandas as pd

auth = tweepy.OAuthHandler(consumer_key, consumer_sectret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)

PLAYERS = pd.read_csv("PlayerInfo.csv", sep = ",", header = 0)
getPlayersTweets(PLAYERS)
```
